```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data_latinoamerica_paises = pd.read_csv(r"C:\Users\USER\Desktop\Avance-Int-M4\PI_Avance_3\DatosFinalesFiltrado.csv")
data_latinoamerica_paises['date'] = pd.to_datetime(data_latinoamerica_paises['date'])
data_latinoamerica_paises.set_index('date', inplace=True)
data_latinoamerica_paises['active_cases'] = data_latinoamerica_paises['cumulative_confirmed'] - data_latinoamerica_paises['cumulative_deceased'] - data_latinoamerica_paises ['cumulative_recovered']

plt.figure(figsize=(25, 20))
sns.lineplot(data=data_latinoamerica_paises.select_dtypes(include=["float64", "int64"]).resample('M').mean(), x='date', y='active_cases', label='Casos Activos')
sns.lineplot(data=data_latinoamerica_paises.select_dtypes(include=["float64", "int64"]).resample('M').mean(), x='date', y='cumulative_recovered', label='Casos Recuperados')
plt.xlabel('´Fecha')
plt.ylabel('Numero de casos')
plt.title('Evolución de casos activos vs. Recuperados en América Latina')
plt.legend()
plt.show()

```
